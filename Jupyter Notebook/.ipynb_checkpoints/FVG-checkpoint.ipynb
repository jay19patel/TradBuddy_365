```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("MainDataSet.csv").tail(1000)
df = df[["Datetime" ,"High","Low","Open","Close"]]
```

```python
import pandas as pd
df['FVG Range'] = "None"
condition_1 = df["Low"].shift(2) - df["High"] > 10
condition_2 = df["Low"] - df["High"].shift(2) > 10

df.loc[condition_1, 'FVG Status'] = "Up Side"
df.loc[condition_2, 'FVG Status'] = "Down Side"

df.loc[condition_1 | condition_2, 'FVG Range'] = df.loc[condition_1 | condition_2, ['High', 'Low']].apply(tuple, axis=1)
df["FVG Gap"] = df['High'] - df['Low']
```

```python
dff = df[df["FVG Range"] != "None"]
dff
```

```python
print("Avrage FVG Gap :",df["FVG Gap"].mean())
```

```python
current_state = []
def FVG(row):
    global current_state
    print("---------------------ROW",row["FVG Range"],"Current State :",current_state)
    if current_state is None and row["FVG Range"] != "None":
        current_state.append(row["FVG Range"])
    else:
        # print("--",current_state , row["High"],row["Low"],row["FVG Status"])
        if current_state[0] >= row["High"] >= current_state[1] or current_state[0] >= row["Low"] >= current_state[1]:
            print("DONE",row["Datetime"])
            current_state = None

dff.apply(FVG,axis=1)
```
