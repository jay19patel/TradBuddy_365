```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("MainDataSet1D.csv").tail(100)
```

```python
df.iloc[-1]
```

```python
df["upper_shadow"] = df['High'] - df[['Open', 'Close']].max(axis=1)
df["lower_shadow"] = df[['Open', 'Close']].min(axis=1) - df['Low']

def is_hammer(row):
    return 1 if row['upper_shadow'] * 2 < row['lower_shadow'] else 0

def is_shooting_star(row):
    return -1 if row['upper_shadow'] > row['lower_shadow'] * 2 else 0

df['Is_Hammer'] = df.apply(is_hammer, axis=1)
df['Is_Shooting_Star'] = df.apply(is_shooting_star, axis=1)
```
